# Métricas de Correlaciones de Pearson y $R^2$

La correlación de Pearson y el coeficiente de determinación R2 son dos métricas estadísticas utilizadas para evaluar diferentes aspectos de la relación entre variables, y aunque están relacionadas, tienen significados distintos.

## Correlación de Pearson
### Definición:

La correlación de Pearson, también conocida como coeficiente de correlación de Pearson, mide la fuerza y la dirección de la relación lineal entre dos variables continuas.
Su valor oscila entre -1 y 1.

### Interpretación:

Un valor de 1 indica una correlación lineal positiva perfecta: a medida que una variable aumenta, la otra también lo hace en una relación perfectamente lineal.
Un valor de -1 indica una correlación lineal negativa perfecta: a medida que una variable aumenta, la otra disminuye en una relación perfectamente lineal.
Un valor de 0 indica que no hay una correlación lineal entre las dos variables.
Fórmula:
$$r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}}$$

## Coeficiente de Determinación $R^2$
### Definición:

El coeficiente de determinación, comúnmente conocido como $R^2$, mide la proporción de la variabilidad en la variable dependiente que es explicada por la variabilidad en la(s) variable(s) independiente(s) en un modelo de regresión.
Su valor oscila entre 0 y 1.
Interpretación:

Un $R^2$ de 1 indica que el modelo de regresión explica toda la variabilidad de los datos de la respuesta alrededor de su media.
Un $R^2$ de 0 indica que el modelo de regresión no explica ninguna de las variaciones de los datos de la respuesta alrededor de su media.
Un $R^2$ más alto significa que el modelo explica mejor la variabilidad de la variable dependiente.
Fórmula:

$$ R^2 = 1 - \frac{\sum (y_i - \hat{y_i})^2}{\sum (y_i - \bar{y})^2} $$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Generar datos sintéticos
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Calcular la correlación de Pearson
r, _ = pearsonr(X.flatten(), y.flatten())
print(f"Correlación de Pearson: {r:.2f}")

# Entrenar un modelo de regresión lineal
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred = lin_reg.predict(X)

# Calcular el coeficiente de determinación (R^2)
r2 = r2_score(y, y_pred)
print(f"Coeficiente de Determinación (R^2): {r2:.2f}")

# Visualizar la relación y la regresión
plt.scatter(X, y, label='Datos')
plt.plot(X, y_pred, color='red', label='Línea de Regresión')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Relación Lineal y Regresión')
plt.show()
```
